```python
foi = 'funcotator_20210306_vcf1.vcf'
allele_freqs, gcs, lociT, bioP = [], [], [], []
for line in open(foi):
    if not line.startswith('#'):
        info_fields = line.strip('\r\n').split('\t')[7].split(';')
        allele_freq = np.nan
        for info_el in info_fields:
            if 'AF' == info_el.split('=')[0]:
                allele_freq = float(info_el.split('=')[1])
            if 'FUNCOTATION' == info_el.split('=')[0]:
                gc_content = float(info_el.split('=')[1].split('|')[19])
                locus_type = info_el.split('=')[1].split('|')[55]
                bio_process = info_el.split('=')[1].split('|')[88]
        allele_freqs.append(allele_freq)
        gcs.append(gc_content)
        lociT.append(locus_type)
        bioP.append(bio_process)
```

```python
np.unique(lociT)
```

```python
rewrite_lociT_dict = {'': 'unknown',
                      'RNA_%2C__%20_long_%20_non-coding': 'lncRNA',
                      'RNA_%2C__%20_micro': 'miRNA',
                      'RNA_%2C__%20_ribosomal': 'rRNA', 
                      'RNA_%2C__%20_small_%20_nuclear' : 'snRNA',
                      'RNA_%2C__%20_small_%20_nucleolar' : 'snoRNA',
                      'complex_%20_locus_%20_constituent' : 'complex locus constituent',
                      'gene_%20_with_%20_protein_%20_product' : 'protein coding gene',
                      'immunoglobulin_%20_gene': 'immunonglobulin gene',
                      'immunoglobulin_%20_pseudogene' : 'immunoglobulin pseudogene',
                      'pseudogene' : 'pseudogene',
                      'readthrough' : 'readthrough',
                      'unknown' : 'unknown'}
lociT = [rewrite_lociT_dict[x] for x in lociT]
```

```python
np.unique(bioP)
```

```python
uniq_lociT = []
n_lociT = []
lociT_bool_dict = {}
for value in np.unique(lociT):
    uniq_lociT.append(value)
    lociT_bool_dict[value] = np.array(lociT) == value
    n_lociT.append(np.sum(lociT_bool_dict[value]))
```

```python
x_s = []
for key in uniq_lociT:
    x_s.append(np.log(np.array(allele_freqs)[lociT_bool_dict[key]]))
print(len(x_s))
```

```python
print(np.amin(allele_freqs), np.amax(allele_freqs), np.mean(allele_freqs))
for i in range(len(x_s)):
    print(np.amin(x_s[i]), np.amax(x_s[i]), np.mean(x_s[i]))
```

```python
new_labels = ["{} (n={})".format(locT, n_val) for locT, n_val in zip(uniq_lociT, n_lociT)]
new_labels
```

```python
fig, ax = plt.subplots()
ax.violinplot(x_s, np.arange(0, len(x_s)*3, 3), showmeans=True, showextrema=False, vert=False)
ax.set_yticks(np.arange(0, len(x_s)*3, 3))
ax.set_yticklabels(new_labels)
#plt.xticks(rotation=80)
ax.set_ylabel('Variant HGNC locus type')
ax.set_xlabel('ln(Allele Frequency)')
plt.tight_layout()
plt.show()
```
